## FP growth algorithm

```python
from mlxtend.frequent_patterns import fpgrowth, association_rules
```

```python
fre_itms = fpgrowth(ndf,
                    min_support = 0.6,
                    use_colnames = True)
fre_itms
```

```python
as_ruels = association_rules(fre_itms,
                             metric = 'lift',
                             min_threshold = 1.5)
as_ruels
```

## Caso 1: Which are the top 10 frequent patterns with high Confidence?

```python
sortConf = as_ruels.sort_values('confidence',
                                ascending = False).head(10)
                                                    
sortConf
```

Which are the top 5 frequent patterns with high confidence?

```python
sortConf = as_ruels.sort_values('confidence',
                                ascending = False).head(5)
                                                    
sortConf
```
